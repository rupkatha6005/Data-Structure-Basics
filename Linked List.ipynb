```python
class Node:
    def __init__(self, e, n):
        self.element = e
        self.next = n
```

```python
class LinkedList:
    def __init__(self, a):
        self.head = None
        if type(a)==list:
            for i in range(len(a)):
                newNode=Node(a[i],None)
                if self.head==None:
                    self.head=newNode
                    tail=newNode
                else:
                    tail.next=newNode
                    tail=newNode
        else:
            self.head=a  
 
    def countNode(self):
        # To Do
        n=self.head
        c=0
        while n!=None:
            c+=1
            n=n.next
        return c

    def printList(self):
        # To Do
        n=self.head
        c=self.countNode()
        for i in range(c):
            if i==c-1:
                print(n.element)
                return
            print(n.element, end=', ')
            n=n.next
    
            
    def nodeAt(self, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        n=self.head
        c=0
        while n!=None:
            if c==idx:
                return n
            n=n.next
            c+=1
        
    def get(self, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        n=self.head
        c=0
        while n!=None:
            if c==idx:
                return n.element
            n=n.next
            c+=1
 
    def set(self, idx, elem):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        n=self.head
        c=0
        while n!=None:
            if c==idx:
                val=n.element
                n.element=elem
                return val
            c+=1
            n=n.next
      
    def indexOf(self, elem):
    # To Do
        n=self.head
        c=0
        while n!=None:
            if elem==n.element:
                return c
            c+=1
            n=n.next
        return -1
  

    def contains(self, elem):
        # To Do
        n=self.head
        c=0
        while n!=None:
            if elem==n.element:
                return True
            c+=1
            n=n.next
        return False
        
        
    def copyList(self):
        # To Do
        n=self.head
        newHead=None
        while n!=None:
            newNode=Node(n.element,None)
            if newHead==None:
                newHead=newNode
                tail=newNode
            else:
                tail.next=newNode
                tail=newNode
            n=n.next
        return newHead 

    def reverseList(self):
        # To Do
        newHead=None
        n=self.head
        while n!=None:
            newNode=Node(n.element,None)
            if newHead==None:
                newHead=newNode
            else:
                newNode.next=newHead
                newHead=newNode
            n=n.next
        return newHead
  
 
    def insert(self, elem, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        newNode=Node(elem,None)
        if idx==0:
            newNode.next=self.head
            self.head=newNode
        else:
            c=0
            n=self.head
            while n!=None:
                if c==idx-1:
                    p=n
                    newNode.next=p.next
                    p.next=newNode
                    break
                c+=1
                n=n.next
                         
    def remove(self, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        n=self.head
        if idx==0:
            q=n
            p=n.next
            self.head=p
        else:
            c=0
            while n!=None:
                if c==idx-1:
                    q=n.next
                    break
                c+=1
                n=n.next
        val=q.element
        n.next=q.next
        q.next=None
        q.element=None
        q=None
        return val
  
    def rotateLeft(self):
        # To Do
        p=self.head
        self.head=p.next
        p.next=None
        n=self.head
        while n.next!=None:
            n=n.next
        n.next=p
  
    def rotateRight(self):
        # To Do
        n=self.head
        while n.next!=None:
            if n.next.next==None:
                p=n
            n=n.next
        n.next=self.head
        p.next=None
        self.head=n

            
```

```python
class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p
```

```python
class DoublyList:
    def __init__(self, a):
        self.head = None
        if type(a)==list:
            self.head=Node(None,None,None)
            self.head.prev=self.head
            self.head.next=self.head
            for i in range(len(a)):
                newNode=Node(a[i],None,None)
                newNode.prev=self.head.prev
                self.head.prev.next=newNode
                newNode.next=self.head
                self.head.prev=newNode
  
    def countNode(self):
        # To Do
        c=0
        n=self.head.next
        while n!=self.head:
            c+=1
            n=n.next
        return c
    
    def forwardprint(self):
        # To Do
        l=self.countNode()
        c=0
        n=self.head.next
        while n!=self.head:
            if c==l-1:
                print(n.element)
                return
            print(n.element,end=', ')
            c+=1
            n=n.next
  
    def backwardprint(self):
        # To Do
        l=self.countNode()
        c=0
        n=self.head.prev
        while n!=self.head:
            if c==l-1:
                print(n.element)
                return
            print(n.element,end=', ')
            c+=1
            n=n.prev
        
  # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        c=0
        n=self.head.next
        while n!=self.head:
            if idx==c:
                return n
            c+=1 
            n=n.next
            
    def indexOf(self, elem):
        # To Do
        c=0
        n=self.head.next
        while n!=self.head:
            if elem==n.element:
                return c
            c+=1
            n=n.next
        return -1

    def insert(self, elem, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        newNode=Node(elem,None,None)
        if idx==self.countNode():
            n=self.head
        else:
            n=self.nodeAt(idx)
        newNode.prev=n.prev
        n.prev.next=newNode
        n.prev=newNode
        newNode.next=n
        return
    
    def remove(self, idx):
        # To Do
        if idx<0 or idx>self.countNode():
            return None
        n=self.nodeAt(idx)
        n.prev.next=n.next
        n.next.prev=n.prev
        val=n.element
        n.element=None
        n.next=None 
        n.prev=None
        n=None
        return str(val)
```
